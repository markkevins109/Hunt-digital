```python
#importing the file and the required python library for data cleaning
import pandas as pd
df=pd.read_excel(r"C:\Users\mark kevin\Downloads\Data Analyst _ Sample Data _HDM.xlsx")
df.head()
```

```python
# Drop duplicates
df.drop_duplicates(inplace=True)
```

```python
numerical_columns = ['Agent Duration(seconds)', 'Customer Duration(seconds)', 'Attempt No']
```

```python
categorical_columns = ['Call Status', 'Lead Status', 'Agent Name', 'Call Status']
```

```python
# Handle missing values
for column in numerical_columns:
    df[column].fillna(df[column].mean(), inplace=True)
```

```python
# Handle missing values
for column in categorical_columns:
    df[column].fillna(df[column].mode()[0], inplace=True)
```

```python
date_columns = ['Created At', 'Updated At']
```

```python
#Change date in to date format
for column in date_columns:
    df[column] = pd.to_datetime(df[column])
```

```python
for column in categorical_columns:
    df[column] = df[column].astype('category')
```

```python
#Remove null values
df['From Calling Number'].fillna('Unknown', inplace=True)
```

```python
#Save the cleaned fine
cleaned_file_path = 'cleaned_data.xlsx'
df.to_excel(cleaned_file_path, index=False)
print(f"\nCleaned data saved to {cleaned_file_path}")
```
