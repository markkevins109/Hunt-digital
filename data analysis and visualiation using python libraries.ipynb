```python
import pandas as pd
df=pd.read_excel(r"C:\Users\mark kevin\cleaned_data.xlsx")
df
```

```python
total_calls = df['Call Id'].count()
total_calls
```

```python
unique_leads = df['Lead Id'].nunique()
unique_leads
```

```python
calls_connected = df[df['Call Status'] == 'completed']['Call Id'].count()
calls_connected
```

```python
unique_calls_connected = df[df['Call Status'] == 'completed']['Lead Id'].nunique()
unique_calls_connected
```

```python
leads_converted = df[df['Lead Status'] == 'Interested']['Call Id'].count()
leads_converted
```

```python
qualified_leads = df[df['Lead Status'] == 'Interested']['Lead Id'].nunique()
qualified_leads
```

```python
leads_lost = df[df['Lead Status'] == 'Not Interested']['Call Id'].count()
leads_lost
```

```python
avg_agent_call_duration = df['Agent Duration(seconds)'].mean()
avg_agent_call_duration
```

```python
avg_customer_call_duration = df['Customer Duration(seconds)'].mean()
avg_customer_call_duration
```

```python
campaign_performance_report = {
    'Total Calls': total_calls,
    'Unique Leads': unique_leads,
    'Calls Connected': calls_connected,
    'Unique Calls Connected': unique_calls_connected,
    'Leads Converted': leads_converted,
    'Qualified Leads': qualified_leads,
    'Leads Lost': leads_lost,
    'Average Agent Call Duration': avg_agent_call_duration,
    'Average Customer Call Duration': avg_customer_call_duration
}
```

```python
print("\nCampaign Performance Report:")
for key, value in campaign_performance_report.items():
    print(f"{key}: {value}")
```

```python
lead_disposition_report = df['Lead Status'].value_counts()
lead_disposition_report
```

```python
print("\nLead Disposition Report:")
print(lead_disposition_report)
```

```python
# Group by agent name to calculate metrics per agent
agent_performance_report = df.groupby('Agent Name').agg(
    total_calls=('Call Id', 'count'),
    calls_connected=('Call Status', lambda x: (x == 'Answered').sum()),
    leads_converted=('Lead Status', lambda x: (x == 'Interested').sum()),
    avg_agent_call_duration=('Agent Duration(seconds)', 'mean'),
    avg_customer_call_duration=('Customer Duration(seconds)', 'mean')
).reset_index()

# Display the Agent Performance Report
print("\nAgent Performance Report:")
print(agent_performance_report)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Campaign Performance Visualization with Values on Bars
plt.figure(figsize=(12, 8))
ax = sns.barplot(x=list(campaign_performance_report.keys()), y=list(campaign_performance_report.values()), palette='viridis')
plt.title('Campaign Performance Metrics')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.xticks(rotation=45)

# Add values on the bars
for i, value in enumerate(campaign_performance_report.values()):
    ax.text(i, value + 0.05 * value, f'{value:.2f}', ha='center', va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Lead Disposition Visualization with values on bars and rotated x-axis labels
plt.figure(figsize=(10, 6))
lead_disposition_plot = sns.barplot(x=lead_disposition_report.index, y=lead_disposition_report.values, color='skyblue')
plt.title('Lead Disposition Distribution')
plt.xlabel('Lead Status')
plt.ylabel('Count')

# Add values on top of the bars
for index, value in enumerate(lead_disposition_report.values):
    plt.text(index, value + 10, str(value), ha='center', va='bottom')

# Rotate x-axis labels for better readability
plt.xticks(rotation=60)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Agent Performance Visualization with values on bars and rotated x-axis labels
plt.figure(figsize=(14, 8))
agent_performance_plot = sns.barplot(x='Agent Name', y='total_calls', data=agent_performance_report, palette='viridis')
plt.title('Total Calls Made by Agents')
plt.xlabel('Agent Name')
plt.ylabel('Total Calls')

# Add values on top of the bars
for index, row in agent_performance_report.iterrows():
    plt.text(index, row['total_calls'] + 5, str(row['total_calls']), ha='center', va='bottom')

plt.xticks(rotation=60)
plt.show()
```

```python
# Group by campaign_id and lead_status to calculate the distribution of call statuses per campaign
campaign_lead_disposition_report = df.groupby(['Campaign ID', 'Lead Status']).size().unstack(fill_value=0)
# Display the Campaign-wise Lead Disposition Report
print("\nCampaign-wise Lead Disposition Report:")
print(campaign_lead_disposition_report)

# Visualize the Campaign-wise Lead Disposition Report
plt.figure(figsize=(12, 8))
campaign_lead_disposition_report.plot(kind='bar', stacked=True)
plt.title('Campaign-wise Lead Disposition')
plt.xlabel('Campaign ID')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Lead Status')
plt.show()
```
